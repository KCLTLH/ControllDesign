```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.loadtxt('FINAL1.csv', delimiter=',')
t_c = data[:, 1]/1e9
y_c = data[:, 2]

ddata = np.squeeze(data[np.nonzero(np.diff(data[:,1])),:]) 
t_d = ddata[:,3]/1e9
y_d = ddata[:,4]
r = ddata[:,5]
d = ddata[:,6]
u = ddata[:,7]

fig, ax = plt.subplots()
plt.step(t_c, y_c, where='post', label = 'Continus')
plt.step(t_d, y_d, where='post', label = 'Discrete' )
plt.step(t_d, r, label='reference', where='post')


plt.title('Step input with PID kp =0.6194, ki =0.1, kd =0.35796. Ts = 0.05s')
plt.xlabel('time')
plt.ylabel('Voltage')
plt.legend( loc='best')

fig.savefig('Step Input0.5.jpg', format='jpg', dpi=3600)
```
